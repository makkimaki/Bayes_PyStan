```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use("ggplot")
np.random.seed(123)
```

```python
p_trans = np.zeros([3,3])
```

```python
# 0:office, 1:rest room, 2:experiment
p_trans[0,0] = 0.1
p_trans[0,1] = 0.2 
p_trans[0,2] = 0.7
p_trans[1,0] = 0.1
p_trans[1,1] = 0.4 
p_trans[1,2] = 0.5
p_trans[2,0] = 0.3
p_trans[2,1] = 0.3 
p_trans[2,2] = 0.4
```

```python
p_trans
```

```python
# Monte Carlo step number
NMCS = 400
# current state
c_state = 0
c_arr = [c_state]
print(c_arr)
```

```python
for i in range(NMCS):
    current = np.random.choice(3, 1, p=p_trans[c_state, :])
    c_state = current[0]
    c_arr.append(c_state)

df = pd.DataFrame(c_arr)
```

```python
c_state = 0
np.random.choice(3,1,p=p_trans[c_state, :])
```
