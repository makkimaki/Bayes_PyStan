```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from scipy.special import gamma
%matplotlib inline
```

```python
plt.style.use("ggplot")
np.random.seed(123)
accept_dict = {}
```

```python
NMCS = 40000
Ndim = 5
accept = 0

for i in range(NMCS):
    x = 2 * np.random.rand(Ndim) - 1.0
    r = (np.sum(x**2)) ** 0.5
    if r <= 1:
        accept += 1

accept_ratio = accept/NMCS
analytical = np.pi **(Ndim/2) / (2**Ndim * gamma(Ndim/2 + 1))
print("Accepti raio: ", accept_ratio)
print("Analytical solution: ", analytical)
accept_dict.update({Ndim:accept_ratio / analytical})

# print(x)
# plt.hist(x, bins=int(np.sqrt(Ndim)))
```

```python
NMCS = 40000
Ndim = np.arange(1,16)

accept_ratio = []
analytical = []

for j in Ndim:
    accept = 0
    for i in range(NMCS):
        x = 2 * np.random.rand(Ndim[j-1]) - 1.0
        r = (np.sum(x**2)) ** 0.5
        if r <= 1:
            accept += 1

    accept_ratio.append(accept/NMCS)
    analytical.append(np.pi **(Ndim[j-1]/2) / (2**Ndim[j-1] * gamma(Ndim[j-1]/2 + 1)))

print("Accepti raio: ", accept_ratio)
print("Analytical solution: ", analytical)
# accept_dict.update({Ndim:accept_ratio / analytical})

# print(x)
# plt.hist(x, bins=int(np.sqrt(Ndim)))

accept_ratio = np.array(accept_ratio)
analytical = np.array(analytical)
```

```python
plt.scatter(Ndim, accept_ratio/analytical)
plt.xlabel("Dimension")
plt.ylabel("Ratio[-]")
```
